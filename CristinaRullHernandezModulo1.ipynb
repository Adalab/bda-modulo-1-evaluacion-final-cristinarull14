```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []
        self.clientes = {}
        self.ventas_totales = 0

#1. método agregar producto

    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}
        self.inventario.append(nuevo_producto)

#2. método ver inventario

    def ver_inventario(self):
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: €{producto['precio']}, Cantidad: {producto['cantidad']}")

#3. método buscar producto

    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                print(f"Nombre: {producto['nombre']}, Precio: €{producto['precio']}, Cantidad: {producto['cantidad']}")
                return
        print("El producto no está en el inventario")

#4. método actualizar stock

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return
        print("El producto no está en el inventario")

#5. método eliminar producto

    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto)
                return
        print("El producto no está en el inventario")

#6. método calcular el valor del inventario

    def calcular_valor_inventario(self):
        valor_total = sum(producto["precio"] * producto["cantidad"] for producto in self.inventario)
        print(f"El valor total del inventario es: €{valor_total}")

#7. método para realizar la compra

    def realizar_compra(self):
        carrito = {}
        while True:
            self.ver_inventario()
            nombre_producto = input("¿Qué producto desea comprar? (salir para finalizar)")
            if nombre_producto.lower() == "salir":
                break
            cantidad = int(input("¿Qué catidad?"))
            for producto in self.inventario:
                if producto["nombre"] == nombre_producto:
                    if producto["cantidad"] >= cantidad:
                        producto["cantidad"] -= cantidad
                        if nombre_producto in carrito:
                            carrito[nombre_producto]["cantidad"] += cantidad
                        else:
                            carrito[nombre_producto] = {"precio": producto["precio"], "cantidad": cantidad}
                    else:
                        print("No hay stock")
                    break
            else:
                print("El producto no está en el inventario")

        total_compra = sum(item["precio"] * item["cantidad"] for item in carrito.values())
        print(f"El importe total de la compra es: €{total_compra}")
        return carrito, total_compra

#8. método para procesar el pago, cálculando el cambio

    def procesar_pago(self, total_compra):
        while True:
            try:
                pago = float(input("Indica la cantidad de pago "))
                if pago >= total_compra:
                    cambio = pago - total_compra
                    print(f"Pago realizado correctamente. El cambio es: €{cambio}")
                    return True
                else:
                    print("Dinero insuficiente. Ponga otra cantidad")
            except ValueError:
                print("Dinero insuficiente. Ponga una cantidad")

#9. método para agregar un cliente

    def agregar_cliente(self, nombre, email):
        if nombre not in self.clientes:
            self.clientes[nombre] = {"email": email, "compras": []}
        else:
            print("Este cliente ya está en la base de datos")

#10. método para ver todos los clientes 

    def ver_clientes(self):
        for nombre, detalles in self.clientes.items():
            print(f"Nombre: {nombre}, Email: {detalles['email']}")


#11. método para registrar la compra al cliente, actualizando las ventas totales y añadiendola al historial del cliente

    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente not in self.clientes:
            print("Este cliente no está en la base de datos")
            return

        total_compra = sum(item["precio"] * item["cantidad"] for item in carrito.values())
        self.clientes[nombre_cliente]["compras"].append({"productos": carrito, "total": total_compra})
        self.ventas_totales += total_compra

#12. método para ver las compras que ha realizado un cliente


    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            compras = self.clientes[nombre_cliente]["compras"]
            for compra in compras:
                print(f"Compra - Total: €{compra['total']}, Productos: {compra['productos']}")
        else:
            print("Este cliente no está en la base de datos")

#13. método para calcular las ventas totales realizadas

    def calcular_ventas_totales(self):
        print(f"Las ventas totales de la tienda son: €{self.ventas_totales}")
   
```
